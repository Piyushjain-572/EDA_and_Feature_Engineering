```python
import numpy as np
import pandas as pd

from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

import seaborn as sns
```

```python
df = pd.read_csv('/kaggle/input/modified-titanic-dataset/train.csv',usecols=['Age','Pclass','SibSp','Parch','Survived'])
df.dropna(inplace=True)
df.head()
```

```python
X = df.iloc[:,1:5]
y = df.iloc[:,0]
```

```python
np.mean(cross_val_score(LogisticRegression(solver='liblinear'),X,y,scoring='accuracy',cv=20))
```

# Applying Feature Construction

```python
X['Family_size'] = X['SibSp'] + X['Parch'] + 1
X.head()
```

```python
def myfunc(num):
    if num == 1:
        #alone
        return 0
    elif num >1 and num <=4:
        # small family
        return 1
    else:
        # large family
        return 2
```

```python
X['Family_type'] = X['Family_size'].apply(myfunc)
X.head()
```

```python
X.drop(columns=['SibSp','Parch','Family_size'],inplace=True)
X.head()
```

```python
np.mean(cross_val_score(LogisticRegression(),X,y,scoring='accuracy',cv=20))
```

# Feature Splitting

```python
df = pd.read_csv('/kaggle/input/modified-titanic-dataset/train.csv')
df.dropna(inplace=True)
df.head()
```

```python
df['Title'] = df['Name'].str.split(', ', expand=True)[1].str.split('.', expand=True)[0]
df['Name'].str.split(', ', expand=True)[1].str.split('.', expand=True)[0]
```

```python
df[['Title','Name']]
```

```python
(df.groupby('Title')['Survived'].mean()).sort_values(ascending=False)
```

```python
df['Is_Married'] = 0
df.loc[df['Title'] == 'Mrs','Is_Married'] = 1
```
