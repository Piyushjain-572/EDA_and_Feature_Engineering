```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split

from sklearn.impute import KNNImputer,SimpleImputer
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('/kaggle/input/modified-titanic-dataset/train.csv',usecols=['Age','Pclass','Fare','Survived'])
df.head()
```

```python
df.isnull().sum()
```

```python
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=['Survived']),
                                                df['Survived'],
                                                test_size=0.2,
                                                random_state=2)
X_train.isnull().sum()
```

```python
knn = KNNImputer(n_neighbors=3,weights='distance')

X_train_trf1 = knn.fit_transform(X_train)
X_test_trf1 = knn.transform(X_test)
```

```python
clf = LogisticRegression()
clf.fit(X_train_trf1,y_train)

y_pred = clf.predict(X_test_trf1)
print(accuracy_score(y_test,y_pred))
```

```python
# Comparing with Simple Imputer
si = SimpleImputer()

X_train_trf2 = si.fit_transform(X_train)
X_test_trf2 = si.transform(X_test)
```

```python
clf = LogisticRegression()
clf.fit(X_train_trf2,y_train)

y_pred = clf.predict(X_test_trf2)
print(accuracy_score(y_test,y_pred))
```
