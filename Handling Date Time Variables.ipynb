```python
import numpy as np
import pandas as pd
```

```python
time_df = pd.read_csv('/kaggle/input/time-data/messages.csv')
date_df = pd.read_csv('/kaggle/input/online-orders/orders.csv')
```

# Handling Date 

```python
date_df.sample(10)
```

```python
date_df.info()
```

### *See, date col. is of Dtype object. We need to convert it to datetime format to be use it.*

```python
# Converting to datetime datatype
date_df['date'] = pd.to_datetime(date_df['date'])

#OR while reading csv use parse_dates param.
```

```python
date_df.info()
```

### Extract Year

```python
date_df['Year'] = date_df['date'].dt.year
date_df.head()
```

### Extract Month

```python
date_df['Month'] = date_df['date'].dt.month
date_df.head()
```

### Extract Month Name

```python
date_df['Month_name'] = date_df['date'].dt.month_name()
date_df.head()
```

### Extract Day

```python
date_df['Day'] = date_df['date'].dt.day
date_df.head()
```

### Extract Day Name

```python
date_df['Day_name'] = date_df['date'].dt.day_name()
date_df.head()
```

### Extract Day Of Week

```python
date_df['DayOfWeek'] = date_df['date'].dt.dayofweek #In this numbering starts from 0 -> Monday.
date_df.head()
```

### Extract Weekend or not

```python
date_df['Weekend?'] = np.where(date_df['Day_name'].isin(['Saturday','Sunday']),1,0)
date_df.drop(columns=['product_id','city_id','orders'],inplace=True) #Since we wont be using them
date_df.head()
```

### Extract Week Number

```python
date_df['Week_Num'] = date_df['date'].dt.strftime('%U').astype(int)
# dt.week <-- Depreceated
date_df.head()
```

### Extract Quarter

```python
date_df['Quarter'] = date_df['date'].dt.quarter
date_df.head()
```

### Date Difference

```python
import datetime
today = datetime.datetime.now()
date_df['Date_difference'] = today - date_df['date']
date_df.head()
```

### Date Difference in days

```python
date_df['Date_difference_days'] = (today - date_df['date']).dt.days
date_df.head()
```

### Date Difference in Months

```python
#date_df['Date_difference_Months'] = ((today - date_df['date']) / np.timedelta64(1,'M')).astype(int) # Giving error
date_df['Date_difference_Months'] = ((today - date_df['date']).dt.days / 30.44).astype(int) #Avg. days in month
date_df.head()
```
