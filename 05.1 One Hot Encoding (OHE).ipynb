# 2. OHE using sklearn

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=['selling_price']),
                                                df['selling_price'],
                                                test_size = 0.2,
                                                random_state = 2)
print(X_train.size)
X_train.sample(5)
```

```python
from sklearn.preprocessing import OneHotEncoder
oe = OneHotEncoder(drop='first',sparse_output=False,dtype=np.int32) 
#sparse_output = false will result in arrays otherwise it will give sparse matrix.

oe.fit(X_train[['fuel','owner']])

X_train_oe = oe.transform(X_train[['fuel','owner']])
X_test_oe = oe.transform(X_test[['fuel','owner']])
```

```python
X_train_array = np.hstack((X_train[['brand','km_driven']].values,X_train_oe))
print(X_train.shape)
print(X_train_array)
```

### **Note**: *Here we have to do so much manual work, first we applied transformation and then converted to array and finally joined with remaining cols of the data using stacking and then again convert to df. This will be eleminated when we study Column Transformer.*

# 3. OHE with Top Categories or Most Frequent Variables
### Eg: brand col.

```python
print(df['brand'].value_counts())
```

```python
counts = df['brand'].value_counts()         #pandas series with count of each brand
threshold = 100
repl = counts[counts<=threshold].index    #name of brands whose count < threshold
df['brand'] = df['brand'].replace(repl,'uncommon')
df.sample(10)
```

```python
# Now apply ohe
pd.get_dummies(df['brand'],drop_first='True')
```
