```python
import numpy as np 
import pandas as pd 
```

```python
df = pd.read_csv('/kaggle/input/customer/customer.csv')
df.sample(5)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=['age','purchased']),
                                                df['purchased'],
                                                test_size = 0.2,
                                                random_state = 2)
print(X_train.size)
X_train.sample(5)
```

# Ordinal Encoding on input features

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories=[['Female','Male'],['Poor','Average','Good'],['School','UG','PG']])
oe.fit(X_train)

X_train_oe = oe.transform(X_train)
X_test_oe = oe.transform(X_test)
```

```python
X_train = pd.DataFrame(X_train_oe,columns=X_train.columns)
X_test = pd.DataFrame(X_test_oe,columns=X_train.columns)
X_train.sample(10)
```

```python
oe.categories_
```

# Label Encoder on target variable

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(y_train)

y_train_le = le.transform(y_train)
y_test_le = le.transform(y_test)
```

```python
y_train = pd.DataFrame(y_train_le,columns=['purchased'])
y_test = pd.DataFrame(y_test_le,columns=['purchased'])
y_train.sample(10)
```
